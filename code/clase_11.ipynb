### Bibliotecas Importadas

### Matriz A

```python
import numpy as np

A = np.array([
    [1, 2],
    [3, 4],
    [5, 6]
])


print(A.shape)
```

### Descomposición SVD de una Matriz

```python
U, S, VT = np.linalg.svd(A)
print(f"Matriz U: {U}")
print(f"Matriz Sigma: {S}")
print(f"Matriz VT: {VT}")
```

### Creación de Matriz Diagonal de Sigma

```python
diagonal_S = np.diag(S)
print(f"Matriz diagonal de Sigma: {diagonal_S}")


filas_ceros=np.zeros((1,2))
print(f"Filas de ceros: {filas_ceros}")

Sigma = np.concatenate((diagonal_S, filas_ceros), axis=0)
print(f"Matriz Sigma con ceros añadidos: {Sigma}")
```

### Reconstrucción de la Matriz A a partir de U, Sigma y VT

```python
print(f"Matriz U: {U}")
print(f"Matriz Sigma: {Sigma}")
print(f"Matriz VT: {VT}")

A_reconstruida = U @ Sigma @ VT
print(f"Matriz A reconstruida: {A_reconstruida}")
```

### Importación de Bibliotecas en Python

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Vector y Matrices Operadoras

```python
v_original=np.array([2,1])
Matriz_rotación=np.array([
    [0,-1],
     [1,0]
     ])
Matriz_escalado=np.array([
    [2,0],
    [0,0.5]
    ])
Matriz_inclinación=np.array([
    [1,1],
    [0,1]
    ])
```

### Transformaciones Lineales

```python
v_rotado=Matriz_rotación.dot(v_original)
print("Vector original:",v_original)
print("Vector rotado:",v_rotado)
```

```python
v_escalado=Matriz_escalado.dot(v_original)
print("Vector original:",v_original)
print("Vector escalado:",v_escalado)
v_inclinado=Matriz_inclinación.dot(v_original)
print("Vector inclinado:",v_inclinado)
```

### Matriz Compuesta (Rotación, Escalado e Inclinación)

```python
Matriz_compuesta=Matriz_inclinación @ Matriz_escalado @ Matriz_rotación
v_compuesto=Matriz_compuesta @ v_original
print("Vector compuesto:",v_compuesto)
```

### Visualización de Transformaciones Lineales

```python
import matplotlib.pyplot as plt

# 1. Definimos la función primero
def graficar_transformaciones(ax, title, V_transformado, v_original):
    # Vector original (línea gris)
    ax.plot([0, v_original[0]], [0, v_original[1]], "--", color="grey", label="Vector original")

    # Vector transformado (flecha roja)
    ax.quiver(0, 0, V_transformado[0], V_transformado[1], angles='xy', scale_units='xy', 
              scale=1, color='r', label='Vector transformado')

    ax.set_title(title)
    ax.grid(True)
    ax.set_xlim(-3, 4)
    ax.set_ylim(-3, 4)
    ax.legend()

# 2. CREAMOS EL LIENZO (Solo una vez)
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(24, 6))
fig.suptitle('Tipos de transformaciones lineales aplicadas a un vector')

# 3. Llamadas a las funciones (Asegúrate de que v_rotado, v_escalado, etc. estén definidos)
graficar_transformaciones(ax1, 'Rotación 90 grados', v_rotado, v_original)
graficar_transformaciones(ax2, 'Escalado', v_escalado, v_original)
graficar_transformaciones(ax3, 'Inclinación', v_inclinado, v_original)

# Composición
matriz_compuesta = Matriz_rotación @ Matriz_escalado @ Matriz_inclinación
v_compuesto = matriz_compuesta @ v_original

graficar_transformaciones(ax4, 'Composición', v_compuesto, v_original)

plt.tight_layout() # Opcional: para que no se amontonen los títulos
plt.show()
```

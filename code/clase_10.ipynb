### Importaci√≥n de Bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.array([
    [2, 1],
    [1, 2]
])

U, S, VT = np.linalg.svd(A)
```

```python
theta = np.linspace(0, 2 * np.pi, 200)
circulo = np.array([np.cos(theta), np.sin(theta)])
print(circulo.shape)
```

```python
print(circulo.T.shape, A.T.shape)
```

```python
elipse = circulo.T @ A.T
```

```python
plt.figure(figsize=(8, 8))
plt.plot(circulo[0, :], circulo[1, :], '--', color='gray', label='Circulo unitario')
plt.plot(elipse[:, 0], elipse[:, 1], color='blue', label='Elipse transformada por A')

eje1 = U[:, 0] * S[0]
eje2 = U[:, 1] * S[1]

plt.quiver(0, 0, eje1[0], eje1[1], angles='xy', scale_units='xy', scale=1, color='r', width=0.015, label=f'Eje 1 (longitud sigma1={S[0]:.2f})')
plt.quiver(0, 0, eje2[0], eje2[1], angles='xy', scale_units='xy', scale=1, color='green', width=0.015, label=f'Eje 2 (longitud sigma2={S[1]:.2f})')

plt.xlim(-4, 4);plt.ylim(-4, 4)
plt.title("Interpretaci√≥n geom√©trica de SVD")
plt.legend()    
plt.grid(True)
plt.show()
```

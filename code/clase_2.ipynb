### Autovalores y Autovectores de A

```python
A=np.array([
    [3, 1],
    [1, 2]
    ])

autovalores, autovectores = np.linalg.eig(A)
print("Autovalores:", autovalores)
print("Autovectores:\n", autovectores)
```

```python
v_1=autovectores[:,0]
v_2=autovectores[:,1]
print("v_1:",v_1)
print("v_2:",v_2)

λ_1=autovalores[0]
λ_2=autovalores[1]
print("λ_1:",λ_1)
print("λ_2:",λ_2)
```

```python
Av_1=A@v_1
Av_2=A@v_2
print("A.v_1:",Av_1)
print("A.v_2:",Av_2)
```

```python
λ_1v_1=λ_1*v_1
λ_2v_2=λ_2*v_2
print("λ_1.v_1:",λ_1v_1)
print("λ_2.v_2:",λ_2v_2)
```
